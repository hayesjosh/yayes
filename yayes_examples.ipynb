```python
artist_list = df['cm_artist'].unique()
artist_id = np.random.choice(artist_list)
print(artist_id)

artist_df = df[df['cm_artist'] == artist_id]
```

```python
plt.figure(figsize=(14,10))
time = artist_df['timestp']
followers = artist_df['followers']
sns.lineplot(time, followers)
plt.xlabel('time',rotation=0, fontsize=16,labelpad=10)
plt.ylabel('y',rotation=0, fontsize=18,labelpad=25)
plt.title("No Yayes", fontsize=18)
# plt.savefig('vis/example4/no_yayes.png')
plt.show()
```

```python
followers_yayes = yayes(artist_df['followers'])
plt.figure(figsize=(14,10))
sns.lineplot(time, followers_yayes, color='green')
plt.xlabel('time',rotation=0, fontsize=16,labelpad=10)
plt.ylabel('y',rotation=0, fontsize=18,labelpad=25)
plt.title("With Yayes", fontsize=18)
# plt.savefig('vis/example4/with_yayes.png')
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.lineplot(time, followers, label='no Yayes')
sns.lineplot(time, followers_yayes, color='green', label='with Yayes')
plt.xlabel('time',rotation=0, fontsize=16,labelpad=10)
plt.ylabel('y',rotation=0, fontsize=18,labelpad=25)
plt.title("Yayes Smoothing", fontsize=18)
plt.legend(fontsize=15)
plt.savefig('vis/example4/both.png')
plt.show()
```
